```python
import pickle
pickle_file_path = "/content/drive/MyDrive/chunks.pkl"
with open(pickle_file_path, "rb") as f:
    loaded_chunks = pickle.load(f)
print("Chunks loaded successfully.")
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

model_name = "sentence-transformers/all-mpnet-base-v2"

embeddings_model = HuggingFaceEmbeddings(model_name=model_name)
```

```python
#from langchain_huggingface import HuggingFaceEmbeddings

# Initialize the embeddings model using the alternative embedding model for consistency.
#embeddings_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")

# Re-embed the existing chunks using the new model.
# This produces a list (or numpy array) of embeddings (for example, each might be 384-dimensional).
embedded_docs = embeddings_model.embed_documents([doc.page_content for doc in loaded_chunks])

# (Optional) Debug print for the first chunk's embedding vector.
print("\nEmbedding Vector Length:", len(embedded_docs[0]))
```

```python
persist_directory = "/content/drive/MyDrive/chroma_db_"
```

```python
from langchain_chroma import Chroma

# Initialize the Chroma vector database.
# - If a collection with the given name exists at the persist directory, it will load it.
# - Otherwise, a new collection will be created.
db = Chroma(
    collection_name="vector_database",
    embedding_function=embeddings_model,
    persist_directory=persist_directory
)

# Optionally, call db.get() to see the current state.
print("Current document IDs in the vector store:", db.get()["ids"])
```

```python
db.get()
```

```python
db.add_documents(loaded_chunks)
```

```python
db.get()
```

```python
db.persist()
```

```python
# Verify that documents were added by checking the document IDs.
updated_ids = db.get()["ids"]
print("Number of documents in vector store:", len(updated_ids))
```
