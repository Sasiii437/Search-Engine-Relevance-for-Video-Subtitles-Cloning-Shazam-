```python
import zipfile
import io

count = 0

def decode_method(binary_data):
    global count
    # Decompress the binary data using the zipfile module
    # print(count, end=" ")
    count += 1
    with io.BytesIO(binary_data) as f:
        with zipfile.ZipFile(f, 'r') as zip_file:
            # Assuming there's only one file in the ZIP archive
            subtitle_content = zip_file.read(zip_file.namelist()[0])

    # Now 'subtitle_content' should contain the extracted subtitle content
    return subtitle_content.decode('latin-1')  # Assuming the content is UTF-8 encoded text
```

```python
df_sample['file_content'] = df_sample['content'].apply(decode_method)

df.head()
```

```python
import pandas as pd
import re


# Function to clean text in a single row
def clean_text(text):
    if isinstance(text, str):  # Check if the input is a string
        # Remove numbers and timestamps (e.g., "1", "00:00:06,000 --> 00:00:12,074")
        text = re.sub(r"\d+:\d+:\d+,\d+\s-->\s\d+:\d+:\d+,\d+", "", text)  # Remove timestamp lines
        text = re.sub(r"\d+", "", text)  # Remove standalone numbers

        # Remove escape characters (\n, \r, etc.)
        text = text.replace("\n", " ").replace("\r", " ")

        # Remove extra spaces
        text = re.sub(r"\s+", " ", text).strip()  # Replace multiple spaces with a single space
    return text

# Apply the cleaning function to the entire column
df_sample['cleaned_text'] = df_sample['file_content'].apply(clean_text)

# 'cleaned_text' column will now contain the cleaned version of your text data
#print(df[['cleaned_text']])  # Display the cleaned text column
```

```python
import pandas as pd
from langchain_text_splitters.sentence_transformers import SentenceTransformersTokenTextSplitter

# Assume df_sample is your DataFrame with 24749 rows and the following columns:
# 'num', 'name', 'content', 'file_content', 'cleaned_text'

# Prepare the list of document contents from the 'cleaned_text' column
doc_contents = df_sample['cleaned_text'].tolist()

# Prepare the corresponding metadata for each row using the remaining columns.
# This metadata will travel with each chunk, so later you can map each chunk back to its full row.
meta_datas = df_sample[['num', 'name', 'content', 'file_content']].to_dict(orient='records')

# Create an instance of the SentenceTransformersTokenTextSplitter with a chunk size of 500 and an overlap of 50 tokens.
st_text_splitter = SentenceTransformersTokenTextSplitter(
    model_name="sentence-transformers/all-mpnet-base-v2",
    chunk_size=500,
    chunk_overlap=50
)

# Generate document chunks. Each element in 'chunks' is a Document object
# that includes a 'page_content' (the text chunk) and the associated metadata.
chunks = st_text_splitter.create_documents(doc_contents, meta_datas)

# Optionally, inspect one of the chunks to verify the content and metadata.
print("Example Chunk:")
print("Content:", chunks[0].page_content)
print("Metadata:", chunks[0].metadata)
```

```python
import pickle

# Define the file path on your Google Drive
pickle_file_path = "/content/drive/MyDrive/chunks.pkl"

# Save the chunks list to a pickle file
with open(pickle_file_path, "wb") as f:
    pickle.dump(chunks, f)

print(f"Chunks successfully saved to {pickle_file_path}")
```

```python
import pickle
pickle_file_path = "/content/drive/MyDrive/chunks.pkl"
with open(pickle_file_path, "rb") as f:
    loaded_chunks = pickle.load(f)
print("Chunks loaded successfully.")
```
